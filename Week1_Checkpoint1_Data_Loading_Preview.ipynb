## 5. Boş (null) dəyərlərin yoxlanılması

Hər sütunda neçə boş dəyər olduğunu və bunun ümumi sətir sayına nisbətdə faizini hesablayıram. Faizlə baxmaq vacibdir, çünki məsələn "2000 boş dəyər" 100,000 sətirlik dataset-də cüzi görünsə də, 3,000 sətirlik dataset-də kritik ola bilər.

```python
null_counts = df.isnull().sum()
null_percent = (null_counts / len(df) * 100).round(2)

null_summary = pd.DataFrame({
    "null_count": null_counts,
    "null_percent": null_percent
}).sort_values("null_count", ascending = False)

null_summary[null_summary["null_count"] > 0]
```

**Qeyd (ilkin, sadəcə fiksasiya — həll Checkpoint 2-də olacaq):**

- `Binanın növü` demək olar ki, tamamilə boşdur (99.8%) — real istifadə imkanı yoxdur, silinməyə namizəddir.
- `featured` (96.9%), `vip` (91.5%), `Torpaq sahəsi` (84.6%), `hour_y` (84.5%) — çox yüksək faizdə boşdur. Bunların bir qismi (`vip`, `featured`) əslində "bəli/xeyr" xarakterli sahələr ola bilər — yəni boş dəyər "bu status yoxdur" mənasını verə bilər (MNAR — sistemli boşluq), sadəcə silinməli deyil.
- `mortgage`/`İpoteka` (67%), `shop_name`/`shop_title` (28.5%), `products_label` (27.9%), `unit_price`/`Mərtəbə` (24.6%), `bill_of_sale` (21%) — orta səviyyəli boşluqlar, bunların da səbəbi araşdırılmalıdır (elanın fərdi şəxs yoxsa əmlak agentliyi tərəfindən yerləşdirilməsi ilə əlaqəli ola bilər).
- `Otaq sayı` (9.3%), `repair`/`Təmir` (19% / 6%) — nisbətən aşağı, amma yenə də diqqətlə doldurulmalı sütunlardır.
- `owner_name`/`owner_title` (0.65%), `description` (0.26%), `unit_price` — çox az boşluq.

Bu sütunların hər biri üçün "niyə boşdur" sualının cavabı Checkpoint 2-də (null strategiyası) ayrıca araşdırılacaq — sadəcə silmək əvəzinə, boşluğun təsadüfi (MCAR) yoxsa sistemli (MNAR) olduğuna baxılacaq.
