## 5. Checking for null values

I calculate how many null values exist in each column and what percentage that represents of the total row count. Looking at the percentage matters, because for example "2,000 missing values" may look negligible in a 100,000-row dataset but could be critical in a dataset with only 3,000 rows.

```python
null_counts = df.isnull().sum()
null_percent = (null_counts / len(df) * 100).round(2)

null_summary = pd.DataFrame({
    "null_count": null_counts,
    "null_percent": null_percent
}).sort_values("null_count", ascending=False)

null_summary[null_summary["null_count"] > 0]
```

**Observation (initial — just noting it here, the actual handling will happen in Checkpoint 2):**

- `Binanın növü` (building type) is almost entirely empty (99.8%) — practically unusable in its current form, a candidate for dropping.
- `featured` (96.9%), `vip` (91.5%), `Torpaq sahəsi` / land area (84.6%), `hour_y` (84.5%) — very high missing rates. Some of these (`vip`, `featured`) are likely boolean-style fields — meaning a missing value probably means "this status does not apply" (MNAR — systematic missingness), not something to simply drop.
- `mortgage`/`İpoteka` (67%), `shop_name`/`shop_title` (28.5%), `products_label` (27.9%), `unit_price`/`Mərtəbə` (floor) (24.6%), `bill_of_sale` (21%) — moderate missing rates; the reason behind these also needs investigation (it could be related to whether the listing was posted by an individual or a real estate agency).
- `Otaq sayı` (number of rooms) (9.3%), `repair`/`Təmir` (renovation) (19% / 6%) — relatively low, but still columns that need to be filled carefully.
- `owner_name`/`owner_title` (0.65%), `description` (0.26%), `unit_price` — very little missingness.

The "why is this missing" question for each of these columns will be investigated in detail in Checkpoint 2 (null-handling strategy) — instead of simply dropping them, I'll examine whether the missingness is random (MCAR) or systematic (MNAR).

## 6. Checking for duplicate rows

First, I check for rows that are fully identical across all columns. Then I check the uniqueness of the identifier columns (`estate_id`, `id_x`, `id_y`) — because even without full-row duplicates, the same listing could still have been recorded more than once at different scrape times.

```python
print("All-column duplicate row count:", df.duplicated().sum())
```

```python
print("Total row count:", len(df))
print("Unique estate_id count:", df["estate_id"].nunique())
print("Unique id_x count:", df["id_x"].nunique())
print("Unique id_y count:", df["id_y"].nunique())
print("Unique rel_url count:", df["rel_url"].nunique(), "(this is the scraped page URL, not a listing ID — being low is expected)")
```

**Observation:**
- Full-row duplicates: **0** — there are no rows in the dataset that are identical across every column.
- `estate_id`, `id_x`, and `id_y` each have 100,775 unique values, matching the total row count — meaning every row represents a distinct property listing, with no duplication at the ID level.
- `rel_url` has only 411 unique values — this is because the column represents the **page** the listing was scraped from during pagination (e.g. `/alqi-satqi?page=174`), not the individual listing itself. This is not a duplication issue; it simply reflects what the column represents.
- An additional point I noticed: the dataset contains paired columns like `id_x`/`id_y`, `estate_rel_url_x`/`estate_rel_url_y`, `estate_details_id_x`/`estate_details_id_y` — this likely comes from merging two separate tables (a listing table + a listing-details table). I'll examine more closely in Checkpoint 2 whether these columns simply duplicate each other's information.
